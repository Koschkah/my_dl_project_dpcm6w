```python
import os
import time
import requests
from pathlib import Path
import pandas as pd
from datetime import datetime
from google.transit import gtfs_realtime_pb2
from dotenv import load_dotenv
print("Import success!")
```

```python
print(Path("/workspace/.env").exists())
# --- c√©l el√©r√©si √∫t a /workspace/data al√° ---
data_dir = Path("/workspace/data")
data_dir.mkdir(parents=True, exist_ok=True)

csv_path = data_dir / "test_file.csv"

# --- p√©ldaadatok ---
df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["bus", "tram", "metro"],
    "speed_kmh": [45, 60, 80]
})

# --- ment√©s ---
df.to_csv(csv_path, index=False)
print(f"‚úÖ Mentve: {csv_path}")

# --- visszaellen≈ërz√©s ---
print("L√©tezik?", csv_path.exists())
print(df)
```

```python

# --- üîπ .env bet√∂lt√©se ---
BASE_DIR="/workspace"
load_dotenv(dotenv_path=BASE_DIR / ".env")

API_KEY = os.getenv("BKK_API_KEY")
if not API_KEY:
    raise RuntimeError("‚ùå API kulcs nem tal√°lhat√≥ (.env f√°jlban: BKK_API_KEY=...)")


DATA_DIR = Path("/data")
DATA_PATH = DATA_DIR / "vehicle_updates.csv"
# --- üîπ API URL ---
URL = f"https://go.bkk.hu/api/query/v1/ws/gtfs-rt/full/VehiclePositions.pb?key={API_KEY}"

print(f"üöç BKK FUT√ÅR adatgy≈±jt√©s indul: {URL}")
print("Ment√©s helye:", DATA_PATH)
print("Nyomj Ctrl+C-t a le√°ll√≠t√°shoz.\n")
i = 4
while i>0:
    try:
        response = requests.get(URL, headers={"Accept": "application/x-protobuf"}, timeout=10)

        if not response.ok:
            print(f"‚ö†Ô∏è HTTP hiba: {response.status_code}")
            time.sleep(60)
            continue

        feed = gtfs_realtime_pb2.FeedMessage()
        feed.ParseFromString(response.content)

        now = datetime.utcnow()
        records = []

        for entity in feed.entity:
            v = entity.vehicle
            records.append({
                "timestamp_utc": now.strftime("%Y-%m-%d %H:%M:%S"),
                "vehicle_id": entity.id,
                "route_id": v.trip.route_id,
                "lat": v.position.latitude,
                "lon": v.position.longitude,
                "speed": getattr(v.position, "speed", None),
                "heading": getattr(v.position, "bearing", None),
                "vehicle_timestamp": datetime.fromtimestamp(v.timestamp)
            })

        df_new = pd.DataFrame(records)
        print(df_new)
        df_new.to_csv(DATA_PATH, mode="a", header=not DATA_PATH.exists(), index=False)

        print(f"‚úÖ {len(records)} rekord mentve {now.strftime('%H:%M:%S')}")

    except Exception as e:
        print("‚ùå Hiba t√∂rt√©nt:", e)

    time.sleep(60)
```
